# Line Plots


```
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn')
import numpy as np
```

```
fig = plt.figure()
ax = plt.axes() # if this line is not there its not displayed
```

```
x = np.linspace(0,10,1000)
ax.plot(x,np.sin(x))
```

```
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
```

```
plt.plot(x, np.sin(x-0), color='blue')
plt.plot(x, np.sin(x-1), color='g')
plt.plot(x, np.sin(x-2), color='r')
plt.plot(x, np.sin(x-3), color='c')
plt.plot(x, np.sin(x-4), color='m')
plt.plot(x, np.sin(x-5), color='y')
plt.plot(x, np.sin(x-5), color='k')
```

```
plt.plot(x,x, linestyle = 'solid')
plt.plot(x,x-1, linestyle = 'dashed')
plt.plot(x,x-2, linestyle = 'dotted')
plt.plot(x,x-3, linestyle = 'dashdot')

plt.plot(x,x+3, '-g')
plt.plot(x,x+4, '--c')
plt.plot(x,x+6, ':y')
plt.plot(x,x+7, '-.k')
```

```
plt.plot(x,np.sin(x))
plt.xlim(0,10)
plt.ylim(-1.2,1.2)
```

```
plt.plot(x,np.sin(x))
plt.xlim(8,4)
plt.ylim(-1.2,1.2)
```

```
plt.plot(x,np.sin(x))
plt.axis([-1,10,-1.9,1.9])
```

```
plt.plot(x,np.sin(x))
plt.axis([0,10,-2,2])
plt.title("Sine Wave")
plt.xlabel("Time")
plt.ylabel('Value')
```

```
plt.plot(x,np.sin(x), '-g',label='sin(x)')
plt.plot(x,np.cos(x),'-r',label='cos(x)')
plt.axis('equal')
plt.legend()
```

```
ax = plt.axes()
ax.plot(x,np.sin(x))
ax.set(xlim=(0,10), ylim=(-2,2),xlabel = 'x', ylabel='sin(x)',title='Simple Plot')
```

# Scatter Plots

```
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```
x = np.linspace(0,10,30)
y=np.sin(x)
plt.plot(x,y,'o',color='b')
```

```
plt.scatter(x,y,marker='o')
```

```
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.randn(100)
sizes = 1000 * rng.randn(100)
plt.scatter(x,y,c=colors,s=sizes, alpha=0.3,cmap='viridis')
plt.colorbar()
```

```
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T
plt.scatter(features[0],features[1],s=100*features[3],c=iris.target,alpha=0.3,cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
```
